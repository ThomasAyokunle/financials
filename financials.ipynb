# data cleaning and visualization for financials.csv

# DATA CLEANING

```python
# import the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the csv file
financial = pd.read_csv('financials.csv')
```

The codes below converts the columns (' Revenue per cookie ',' Cost per cookie ',' Revenue ',' Cost ',' Profit ') datatype to string, replaces the '$' and ',' sign, strips the whitespaces and then converts the datatype to float

```python
financial[' Revenue per cookie '] = financial[' Revenue per cookie '].astype(str)
financial[' Revenue per cookie '] = financial[' Revenue per cookie '].str.replace('$', '')
financial[' Revenue per cookie '] = financial[' Revenue per cookie '].str.strip()
financial[' Revenue per cookie '] = financial[' Revenue per cookie '].astype(float)
```

```python
financial[' Cost per cookie '] = financial[' Cost per cookie '].astype(str)
financial[' Cost per cookie '] = financial[' Cost per cookie '].str.replace('$', '')
financial[' Cost per cookie '] = financial[' Cost per cookie '].str.strip()
financial[' Cost per cookie '] = financial[' Cost per cookie '].astype(float)
```

```python
financial[' Revenue '] = financial[' Revenue '].astype(str)
financial[' Revenue '] = financial[' Revenue '].str.replace('$', '')
financial[' Revenue '] = financial[' Revenue '].str.replace(',', '')
financial[' Revenue '] = financial[' Revenue '].str.strip()
financial[' Revenue '] = financial[' Revenue '].astype(float)
```

```python
financial[' Cost '] = financial[' Cost '].astype(str)
financial[' Cost '] = financial[' Cost '].str.replace('$', '')
financial[' Cost '] = financial[' Cost '].str.replace(',', '')
financial[' Cost '] = financial[' Cost '].str.strip()
financial[' Cost '] = financial[' Cost '].astype(float)
```

```python
financial[' Profit '] = financial[' Profit '].astype(str)
financial[' Profit '] = financial[' Profit '].str.replace('$', '')
financial[' Profit '] = financial[' Profit '].str.replace(',', '')
financial[' Profit '] = financial[' Profit '].str.strip()
financial[' Profit '] = financial[' Profit '].astype(float)
```

# DATA VISUALIZATION

```python
country_by_units_sold = financial.groupby(['Country'])['Units Sold'].sum()
```

```python
result = financial.groupby('Country').aggregate({'Units Sold' : 'sum', ' Cost ':'sum',' Profit ':'sum'})
result.plot.bar()
plt.title("Units Sold, Cost and Profit by Country and Year")
plt.xlabel("Country")
```

```python
results = financial.groupby(['Country', 'Year']).aggregate({' Cost ': 'sum', ' Profit ': 'sum'})
results_sorted = results.sort_values(by=[' Cost '])
results_sorted.plot.barh()
plt.title("Cost and Profit by Country and Year")
```

```python
results = financial.groupby(['Country', 'Year']).aggregate({' Revenue ': 'sum'})
results_sorted = results.sort_values(by=' Revenue ', ascending= True)
results_sorted.plot.barh()
plt.title("Revenue by Country and Year")
plt.xlabel("Revenue")
```

```python
results = financial.groupby(['Country', 'Year']).aggregate({'Units Sold': 'sum'})
results_sorted = results.sort_values(by='Units Sold', ascending= True)
results_sorted.plot.barh()
plt.title("Units Sold by Country and Year")
plt.xlabel("Units Sold")
```

```python
results = financial.groupby(['Year', ' Month Name ']).aggregate({' Profit ': 'mean'})
results_sorted = results.sort_values(by=' Profit ', ascending= True)
results_sorted.plot.barh()
plt.title("Profit by Year and Month")
plt.xlabel("Profit")
```

```python
results = financial.groupby([' Month Name ', 'Year']).aggregate({' Revenue ': 'sum'})
results_sorted = results.sort_values(by=' Revenue ', ascending= True)
results_sorted.plot.barh()
plt.title("Revenue by Month and Year")
plt.xlabel("Revenue")
```

```python
result = financial.groupby(' Product ').aggregate({'Units Sold' : 'sum'})
result.plot.bar(color= 'c')
plt.title("Units Sold by Product")
plt.ylabel("Units Sold")
```

```python
result = financial.groupby(' Product ').aggregate({' Revenue ' : 'sum'})
result.plot.bar(color = 'c')
plt.title("Revenue by Product")
plt.ylabel("Revenue")
```

```python
result = financial.groupby(' Product ').aggregate({' Profit ' : 'sum', ' Cost ' : 'sum'})
result.plot.bar(color = 'b' 'g')
plt.title("Cost and Profit by Product")
```
